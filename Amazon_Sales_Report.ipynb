```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df.dropna(inplace=True)
```

```python
df['ship-postal-code'] = df['ship-postal-code'].astype('int')
```

```python
df['ship-postal-code'].dtype
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.rename(columns={'Qty' : 'Quantity'})
```

```python
df.describe()
```

```python
df.describe(include='object')
```

```python
df[['Qty','Amount']].describe()
```

```python
ax = sns.countplot(x='Size',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

From the above graph we can see that most of the people ordered M Size.

```python
s_qty = df.groupby('Size', as_index= False)['Qty'].sum().sort_values(by='Qty',ascending=False)
sq = sns.barplot(x='Size', y='Qty', data=s_qty)

for bars in sq.containers:
    sq.bar_label(bars)
```

From the above graph we can see that most quantity ordered is M-Size

```python
plt.figure(figsize=(20,10))
ax = sns.countplot(data=df, x='Courier Status', hue='Status')
for bars in ax.containers:
    ax.bar_label(bars)

plt.show()
```

From the above graph we can see that most the orders are Delievered to the Buyer succefully.

```python
df['Size'].hist()
```

```python
plt.hist(df['Category'], bins=15, edgecolor='Black')
plt.xticks(rotation=90)
plt.show()
```

From the above graph we can see that most people bought T-shirt.

```python
b2b_c = df['B2B'].value_counts()
plt.pie(b2b_c, labels=b2b_c.index, autopct='%1.2f%%')
plt.show()
```

From the above graph we can see that most buyers are the retailers.

```python
plt.scatter(x=df['Category'], y=df['Size'])
plt.xlabel('Category')
plt.ylabel('Size')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df, x='ship-state')
plt.xticks(rotation=90)
plt.xlabel('Ship-state')
plt.ylabel('count')
plt.show()
```

```python
top10 = df['ship-state'].value_counts().head(10)

plt.figure(figsize=(12,6))
sns.countplot(data=df[df['ship-state'].isin(top10.index)], x='ship-state')
plt.xticks(rotation=45)
plt.xlabel('Ship-state')
plt.ylabel('count')
plt.show()
```

From the above graph we can see that most buyers are from Maharastra.
