```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds = pd.read_csv(r'D:\Data Analytics\Diwali Sales Data.csv', encoding='unicode_escape')
```

```python
ds.shape
```

```python
ds.head(10)
```

```python
ds.info()
```

```python
ds.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(ds)
```

```python
pd.isnull(ds).sum()
```

```python
ds.dropna(inplace=True)
```

```python
ds['Amount'] = ds['Amount'].astype('int')
```

```python
ds['Amount'].dtypes
```

```python
ds.describe()
```

```python
ds[['Age', 'Orders', 'Amount']].describe()
```

```python
ds.columns
```

```python
dl = sns.countplot(x='Gender', data=ds)
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
ds.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_gen = ds.groupby(['Gender'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_gen, x='Gender', y='Amount')
```

From the above graph we can see that most of the buyers are Female and even purchasing power of females are greater than men.

```python
dl = sns.countplot(data=ds, x='Age Group', hue='Gender')
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
ds.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_age = ds.groupby(['Age Group'], as_index=False) ['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(data=sales_age, x='Age Group', y='Amount')
```

From above graph we can see that the most of the buyers are of age group between 26-35 yrs female.

```python
sales_state = ds.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
dl = sns.barplot(data=sales_state, x='State', y='Orders')
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
sales_state = ds.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(16,5)})
sns.barplot(data=sales_state, x='State', y='Amount')
```

From above graph we can see that most of the orders and total sale amount are from Uttar Pradesh, Maharastra and Karnataka respectively.

```python
dl = sns.countplot(data=ds,x='Marital_Status')
sns.set(rc={'figure.figsize':(4,5)})
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
ds.groupby(['Marital_Status'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sales_status = ds.groupby(['Marital_Status', 'Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data=sales_status, x='Marital_Status', y='Amount', hue='Gender')
```

From above graph we can see that most of the buyers are married (woman) and they have higher purchasing power.

```python
dl = sns.countplot(data=ds,x='Occupation')
sns.set(rc={'figure.figsize':(25,5)})
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
sales_occ = ds.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.set(rc={'figure.figsize':(25,5)})
sns.barplot(data=sales_occ, x='Occupation', y='Amount')
```

From above graph we can see that most of the buyers are working in IT, Healthcare and Aviation Sector.

```python
dl = sns.countplot(data=ds,x='Product_Category')
sns.set(rc={'figure.figsize':(25,5)})
for bars in dl.containers:
    dl.bar_label(bars)
```

```python
sales_product = ds.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data=sales_product, x='Product_Category', y='Amount')
```

From the above graph we can see that most of the sold products are Food, Clothing & Apparel and Electronics & Gadgets category.

# Conclusion :
Married Woman age group 26-35 years from Uttar Pradesh, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing and Electronic Category.
